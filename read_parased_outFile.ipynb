```python
# lets put this into a func

def parseFunc(path):
    df = pd.read_csv(path, sep='\t')
    df['Cell'] = df['Cell'].str.strip('&')
    df['Cell'] = df['Cell'].str.split('@', expand=True)[0]
    
    return(df)
```

```python
files = os.listdir('/Users/lincoln.harris/code/STAR-Fus_scratch/breakdown/')
big_df = pd.DataFrame()

for f in files:
    runName = f.split('_')[0]
    print(runName)
    basePath = '/Users/lincoln.harris/code/STAR-Fus_scratch/breakdown/'
    fullPath = basePath + f
    curr_df = parseFunc(fullPath)
    curr_df['RunName'] = runName
    big_df = pd.concat([big_df, curr_df])
```

```python
big_df
```

```python
# make a new df, with combined JunctionReadCount & SpanningFragCount for each Cell/FusionName
#      this has to be the most inefficient way of doing this...
#           aka monkeySort

dummyRows = {'#FusionName':['hello','world','foo'], 'Cell':['hello','world','foo'], 'JunctionReadCount':[0,0,0], 'SpanningFragCount':[0,0,0], 'RunName':[0,0,0]}
new_df = pd.DataFrame(data=dummyRows)

for i in range(0,len(big_df.index)):
    currRow = big_df.iloc[i]
    currCell = currRow['Cell']
    currFusion = currRow['#FusionName']
    currJctCount = currRow['JunctionReadCount']
    currSpnFragCount = currRow['SpanningFragCount']
    
    toAddRow = currRow[['#FusionName','Cell','JunctionReadCount','SpanningFragCount','RunName']]
    
    # double select method -- does the entry with currFusion and currCell already exist in new_df? 
    sub_df = new_df.loc[(new_df['#FusionName'] == currFusion) & (new_df['Cell'] == currCell)]
    if len(sub_df.index) == 0: # new row to add!!
        new_df = new_df.append(toAddRow)
    else: # update JunctionReadCount and SpanningFragCount
        idx = sub_df.index[0]
        new_df['JunctionReadCount'][idx] += currJctCount
        new_df['SpanningFragCount'][idx] += currSpnFragCount
```

```python
# rearrange cols
new_df = new_df[['RunName','Cell','#FusionName','JunctionReadCount','SpanningFragCount']]
new_df
```

```python
# tweak colnames, slightly
new_df.columns = ['RunName','Cell','Fusion','JunctionReads','SpanningFragReads']
new_df
```

```python
# get rid of dummy rows
new_df = new_df[new_df.Fusion != 'hello']
new_df = new_df[new_df.Fusion != 'world']
new_df = new_df[new_df.Fusion != 'foo']
new_df
```
